```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris_dataset= load_iris()
```

```python
iris_dataset.keys()
```

```python
X_full = iris_dataset['data']
y_full = iris_dataset['target']
feature_names = iris_dataset['feature_names']
target_names = iris_dataset['target_names']
```

```python
#printing feature names
feature_names
```

```python
#printing target names
target_names
```

```python
# we have four features in the dataset
#lets choose 2 features for normalization
```

```python
#selecting the first two features from the 2D array
selected_features= X_full[:,:2]
selected_features
```

```python
#create a dataframe and view the mean ,variance and quartiles
iris_df= pd.DataFrame(data=selected_features, columns=feature_names[:2])
iris_df.describe()
```

```python
#applying normalizations
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, PowerTransformer

minmaxscaler= MinMaxScaler()
normalized_minmaxscaler = minmaxscaler.fit_transform(selected_features)

standardscaler =StandardScaler()
normalized_standardscaler =standardscaler.fit_transform(selected_features)

robust_scaler = RobustScaler()
normalized_robust = robust_scaler.fit_transform(selected_features)

power_transformer = PowerTransformer()
normalized_power = power_transformer.fit_transform(selected_features)
```

```python
selected_features
```

```python
#plotting two features on two histograms
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(data=selected_features, x=selected_features[:,0],  label="Sepal Length") #
sns.histplot(data=selected_features, x=selected_features[:, 1], label="Sepal Width") #
plt.legend()
plt.title("Histograms of Original Data")
plt.xlabel("Feature Values")
plt.ylabel("Frequency")

plt.subplot(1, 2, 2)
sns.histplot(data=normalized_minmaxscaler , x=normalized_minmaxscaler[:,1], label='sepal width')
sns.histplot(data=normalized_minmaxscaler , x=normalized_minmaxscaler[:,0], label='sepal length')
plt.title('after normalizing')
plt.xlabel('feature values')
plt.ylabel('frequency')
plt.legend()
```
